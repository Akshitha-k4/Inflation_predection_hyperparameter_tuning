```python
# importting Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading and reading dataset
df = pd.read_csv("inflation_data_set.csv")
df.head()
```

```python
df.info()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.info()
```

```python
df=df.drop(['Country Code', 'IMF Country Code'], axis=1)
df.head()
```

```python
# Replace all NaN elements with 0s
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Country'] = label_encoder.fit_transform(df['Country'])
df.head()
```

```python
import seaborn as sns # importing seaborn for statistical plots 
import matplotlib.pyplot as plt  # matplotlib.pyplot plots data
```

```python
# Train test split 
from sklearn.model_selection import train_test_split
```

```python
from sklearn import preprocessing
from scipy.stats import zscore
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error
```

```python
X = df.drop( ['2022'] , axis=1)
```

```python
y = df['2022']
```

```python
#train-test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=123)
```

# Linear Regression

```python
lr = LinearRegression() 
lr.fit(X_train, y_train) # Fitting the model on Training data 
y_pred_lr = lr.predict(X_test)
print('\nLR R2_score on training data:',lr.score(X_train, y_train))
print('LR R2_score on test data:',lr.score(X_test, y_test))
print('LR Mean Square Error :',mean_squared_error(y_test,y_pred_lr)) 
print('LR Root Mean Square Error :',np.sqrt(mean_squared_error(y_test, y_pred_lr))) 
print('LR Mean Absolute Error :',mean_absolute_error(y_test, y_pred_lr))
```

# LASSO Regression

```python
# Creating a regularized LASSO model 
lasso = Lasso(alpha=0.1)
lasso.fit(X_train,y_train) # Fitting the model on Training data 
y_pred_lasso = lasso.predict(X_test) # Making predictions on Test data
print("Lasso model coefficients:", (lasso.coef_)) # Printing the coefficients 
print('\nLasso R2_score on training data:',lasso.score(X_train, y_train))
print('Lasso R2_score on test data:',lasso.score(X_test, y_test))
print('Lasso Mean Square Error :',mean_squared_error(y_test,y_pred_lasso)) 
print('Lasso Root Mean Square Error :',np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
print('Lasso Mean Absolute Error :',mean_absolute_error(y_test, y_pred_lasso))
```

# RIDGE Regression

```python
# Creating a regularized RIDGE model
ridge = Ridge(alpha=.3) 
ridge.fit(X_train,y_train) # Fitting the model on Training data
y_pred_ridge = ridge.predict(X_test) # Making predictions on Test data
print("Ridge model coefficients:", (ridge.coef_)) # Printing the coefficients 
print('\nRidge R2_score on training data:',ridge.score(X_train, y_train))
print('Ridge R2_score on test data:',ridge.score(X_test, y_test))
print('Ridge Mean Square Error :',mean_squared_error(y_test,y_pred_ridge)) 
print('Ridge Root Mean Square Error :',np.sqrt(mean_squared_error(y_test, y_pred_ridge)))
print('Ridge Mean Absolute Error :',mean_absolute_error(y_test, y_pred_ridge))
```

# ElasticNet Regression

```python
from sklearn.linear_model import ElasticNet
# Fitting a Elastic Net Regression
en = ElasticNet(alpha=1.0, l1_ratio=0.5)
en.fit(X_train, y_train) # Fitting the model on Training data
y_pred_en = en.predict(X_test) # Making predictions on Test data
print("Elastic Net Regression model coefficients:", (en.coef_)) # Printing the coefficients
print('\nEN R2_score on training data:',en.score(X_train, y_train))
print('EN R2_score on test data:',en.score(X_test, y_test))
print('EN Mean Square Error :',mean_squared_error(y_test,y_pred_en))
print('EN Root Mean Square Error :',np.sqrt(mean_squared_error(y_test, y_pred_en)))
print('EN Mean Absolute Error :',mean_absolute_error(y_test, y_pred_en))
```

# KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Select the feature and target columns
X = df.drop( ['2022'] , axis=1).values
y = df['2022'].values

# Split the data into training and test sets
train_data, test_data, train_target, test_target = train_test_split(X, y, test_size=0.3, random_state=123)

# Perform KNN regression using k = 3 and feature_column as the covariate
knn_reg = KNeighborsRegressor(n_neighbors=3, weights='uniform')
knn_reg.fit(train_data, train_target)

# Predict the target values for the test set
test_pred = knn_reg.predict(test_data)

# Calculate the mean squared error and R^2
mse_knnr = mean_squared_error(test_target, test_pred)
r_sq = r2_score(test_target, test_pred)

# Print the results
print("Mean Squared Error:", mse_knnr)
print("R^2:", r_sq)
```

# Support Vector Regressor

```python
from sklearn.svm import SVR

# Perform SVR
svr = SVR(kernel='linear', C=1, epsilon=0.2)
svr.fit(train_data, train_target)

# Predict the target values for the test set
test_pred = svr.predict(test_data)

# Calculate the mean squared error and R^2
mse_svr = mean_squared_error(test_target, test_pred)
r_sq = r2_score(test_target, test_pred)

# Print the results
print("Mean Squared Error (SVR):", mse_svr)
print("R^2 (SVR):", r_sq)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

# Split the data into training and test sets
train_data, test_data, train_target, test_target = train_test_split(X, y, test_size=0.3, random_state=123)

# Perform Decision Tree Regression
dt_reg = DecisionTreeRegressor()
dt_reg.fit(train_data, train_target)

# Predict the target values for the test set
test_pred = dt_reg.predict(test_data)

# Calculate the mean squared error and R^2
mse_dtr = mean_squared_error(test_target, test_pred)
r_sq = r2_score(test_target, test_pred)

# Print the results
print("Mean Squared Error (Decision Tree):", mse_dtr)
print("R^2 (Decision Tree):", r_sq)
```

#  Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Split the data into training and test sets
train_data, test_data, train_target, test_target = train_test_split(X, y, test_size=0.3, random_state=123)

# Perform Random Forest Regression
rf_reg = RandomForestRegressor()
rf_reg.fit(train_data, train_target)

# Predict the target values for the test set
test_pred = rf_reg.predict(test_data)

# Calculate the mean squared error and R^2
mse_rfr = mean_squared_error(test_target, test_pred)
r_sq = r2_score(test_target, test_pred)

# Print the results
print("Mean Squared Error (Random Forest):", mse_rfr)
print("R^2 (Random Forest):", r_sq)
```

# Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
# Split the data into training and test sets
train_data, test_data, train_target, test_target = train_test_split(X, y, test_size=0.3, random_state=123)
# Perform Gradient Boosting Regression
gb_reg = GradientBoostingRegressor(learning_rate=0.1)
gb_reg.fit(train_data, train_target)

# Predict the target values for the test set
test_pred_gb= gb_reg.predict(test_data)
# Calculate the mean squared error and R^2
mse_gbr = mean_squared_error(test_target, test_pred_gb)
print(mse_gbr)
```

# RandomSearchCV

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.5, 1.0],
    'max_features': [1.0, 'sqrt', 'log2']  # Explicitly set max_features=1.0
}

# Perform RandomizedSearchCV with GradientBoostingRegressor
gb_reg_random = RandomizedSearchCV(estimator=GradientBoostingRegressor(), param_distributions=param_grid, n_iter=100, cv=5, verbose=2, random_state=123)
gb_reg_random.fit(X_train, y_train)

# Predict the target values for the test set
test_pred_rscv = gb_reg_random.predict(X_test)

# Calculate the mean squared error and R^2
mse_gbr = mean_squared_error(y_test, test_pred_rscv)
r_sq = r2_score(y_test, test_pred_rscv)

# Print the best parameters and the best score
print("Best Parameters:", gb_reg_random.best_params_)
print("Best Score:", gb_reg_random.best_score_)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Create a new GradientBoostingRegressor model with the best parameters
gbr = GradientBoostingRegressor(
    subsample=0.5,
    n_estimators=200,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features=1.0,
    max_depth=20,
    learning_rate=0.1
)

# Fit the model to the training data
gbr.fit(train_data, train_target)

# Predict the target values for the test set
test_pred = gbr.predict(test_data)

# Calculate the mean squared error and R^2
mse = mean_squared_error(test_target, test_pred)
r_sq = r2_score(test_target, test_pred)

# Print the results
print("Mean Squared Error (Gradient Boosting):", mse)
print("R^2 (Gradient Boosting):", r_sq)
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [200],
    'learning_rate': [0.1],
    'max_depth': [20],
    'min_samples_split': [2],
    'min_samples_leaf': [1],
    'max_features': [1.0],
    'subsample': [0.5]
}

# Initialize the Gradient Boosting Regressor
gb_reg = GradientBoostingRegressor()

# Perform Grid Search CV
grid_search = GridSearchCV(gb_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(train_data, train_target)

# Print the best parameters and the corresponding mean squared error
print("Best Parameters:", grid_search.best_params_)
print("Mean Squared Error (Gradient Boosting):", -grid_search.best_score_)
```
